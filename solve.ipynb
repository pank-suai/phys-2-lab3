```kotlin
// Функция утилита
fun Double.round(decimals: Int): Double {
    var multiplier = 1.0
    repeat(decimals) { multiplier *= 10 }
    return round(this * multiplier) / multiplier
}
```

#### Задание 2

```kotlin
val df2 = DataFrame.readCSV("./data/2.csv").add("alpha2_avg"){
    ("alpha21"<Int>() + "alpha22"<Int>())/2.0
}
df2
```

```kotlin
val alpha2_avg = df2.alpha2_avg[0]
alpha2_avg
```

```kotlin
val epsilon0 = ((2 * K * R * Havg * tan(alpha2_avg * PI / 180.0) ) / (N * nu * epsilon * U)).round(13)
epsilon0
```

```kotlin
val с = 1.0/sqrt(epsilon0 * mu0)
val roundC = (с * 10.0.pow(-8)).round(2) // * 10^8
roundC
```

#### Вычисление погрешностей

##### Систематическая погрешность

```kotlin
val thetaU = 0.375 // В
val thetaI = 2 * 10.0.pow(-3)
val thetaAlpha = 0.5 * PI / 180.0
```

```kotlin
val thetaH = (Havg * ((2 * thetaAlpha)/sin(2.0 * results.mean("alpha_avg") * PI / 180.0 ) + thetaI/df.mean("I"))).round(2)
thetaH
```

```kotlin
val thetaEpsilon = (epsilon0 * (thetaH/Havg + thetaAlpha/alpha2_avg + thetaU/U)).round(15)
thetaEpsilon
```

```kotlin
val thetac = (roundC * (thetaEpsilon/epsilon0))
thetac
```

```kotlin
import org.apache.commons.csv.CSVFormat

df2.add("epsilon0") { "${(epsilon0 * 10.0.pow(11)).round(2)} plus.minus ${(thetaEpsilon * 10.0.pow(11)).round(2)}"  }.add("c") {
    "$roundC plus.minus ${thetac.round(2)}"
}.apply { writeCSV("./data/2_res.csv", format = CSVFormat.DEFAULT) }
```
